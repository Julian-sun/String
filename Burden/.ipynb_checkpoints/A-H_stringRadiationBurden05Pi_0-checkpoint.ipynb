```python
def Gn(theta,phi,psi,Mpp,Mmm,n,n0):
    Rn = np.sqrt(1.0-(n0/n)**2)
    mP = n/Mpp
    mM = n/Mmm
    rx = np.cos(theta)
    ry = np.sin(theta)*np.cos(phi)
    rz = np.sin(theta)*np.sin(phi)
    aP = 1
    bP = 0
    aM = np.cos(psi)
    bM = np.sin(psi)
    kP = ry*np.cos(psi)+rz*np.sin(psi)
    kM = ry
    if kP>=0:
        dP = Rn/Mpp*np.sqrt(np.cos(theta)**2+np.sin(theta)**2*np.cos(phi-psi)**2)
    else:
        dP = -Rn/Mpp*np.sqrt(np.cos(theta)**2+np.sin(theta)**2*np.cos(phi-psi)**2)
    if kM>=0:
        dM = Rn/Mmm*np.sqrt(np.cos(theta)**2+np.sin(theta)**2*np.cos(phi)**2 )
    else:
        dM = -Rn/Mmm*np.sqrt(np.cos(theta)**2+np.sin(theta)**2*np.cos(phi)**2 )
    if kP>=0 and kP<0.00000001:
        deltaP = np.pi/2
    elif kP<0 and kP>-0.00000001:
        deltaP = -np.pi/2
#    elif kP==0:
#        deltaP = 0
    else: deltaP = np.arctan(rx/kP)
    if kM>=0 and kM<0.00000001:
        deltaM = np.pi/2
    elif kM<0 and kM>-0.00000001:
        deltaM = -np.pi/2
#    elif kM==0:
#        deltaM = 0
    else: deltaM = np.arctan(rx/kM)
    Jp = sc.special.jv(mP,mP*dP)
    Jm = sc.special.jv(mM,mM*dM)
    JpPr = sc.special.jvp(mP,mP*dP, 1)
    JmPr = sc.special.jvp(mM,mM*dM, 1)
    aT = np.sin(deltaP)/(dP)*Jp
    alT = JpPr*np.cos(deltaP)
    bT = np.cos(deltaP)/(dP)*Jp
    beT = -JpPr*np.sin(deltaP)
    cT = 0#np.cos(deltaP)/(dP)*Jp*bP
    gaT = 0#-JpPr*np.sin(deltaP)*bP
    dT = np.sin(deltaM)/(dM)*Jm
    deT = JmPr*np.cos(deltaM)
    eT = np.cos(deltaM)/(dM)*Jm*aM
    epT = -JmPr*np.sin(deltaM)*aM
    wT = np.cos(deltaM)/(dM)*Jm*bM
    omT = -JmPr*np.sin(deltaM)*bM
    Imag = rz*(eT*alT-dT*beT-bT*deT+aT*epT) + ry*(-wT*alT+dT*gaT+cT*deT-aT*omT) + rx*(wT*beT-eT*gaT-cT*epT+bT*omT)
    Rea = -cT*eT*rx + cT*dT*ry - bT*dT*rz + aT*eT*rz + bT*rx*wT - aT*ry*wT + rz*beT*deT - \
           ry*gaT*deT - rz*alT*epT + rx*gaT*epT + ry*alT*omT - rx*beT*omT
    M1 = (aT**2 + bT**2 + cT**2 + alT**2 + beT**2 + gaT**2)*\
         (dT**2*rx**2 + eT**2*ry**2 + 2*eT*ry*rz*wT + 2*dT*rx*(eT*ry+rz*wT) + \
          (rx*deT+ry*epT)**2 + rz*(2*(rx*deT+ry*epT)*omT+rz*(wT**2+omT**2))   )
    M2 = (dT**2 + eT**2 + wT**2 + deT**2 + epT**2 + omT**2)*\
        (aT**2*rx**2 + bT**2*ry**2 + 2*bT*ry*rz*cT + 2*aT*rx*(bT*ry+rz*cT) + \
          (rx*alT+ry*beT)**2 + rz*(2*(rx*alT+ry*beT)*gaT+rz*(cT**2+gaT**2))   )
    M3 = 2.0*((aT*dT + bT*eT + cT*wT + alT*deT + beT*epT + gaT*omT)* \
        (cT*dT*rx*rz + cT*eT*ry*rz + cT*rz**2*wT + aT*rx*(dT*rx + eT*ry + rz*wT) + \
         bT*ry*(dT*rx + eT*ry + rz*wT) + rx**2*alT*deT + rx*ry*beT*deT + rx*rz*gaT*deT + \
         rx*ry*alT*epT + ry**2*beT*epT + ry*rz*gaT*epT + rx*rz*alT*omT + ry*rz*beT*omT + rz**2*gaT*omT ) - \
        (-dT*alT - eT*beT - wT*gaT + aT*deT + bT*epT + cT*omT)*\
        (rx*rz*wT*alT + ry*rz*wT*beT + rz**2*wT*gaT + dT*rx*(rx*alT + ry*beT + rz*gaT) + \
         eT*ry*(rx*alT + ry*beT + rz*gaT) - aT*rx**2*deT - bT*rx*ry*deT - cT*rx*rz*deT - \
         aT*rx*ry*epT - bT*ry**2*epT - cT*ry*rz*epT - aT*rx*rz*omT - bT*ry*rz*omT - cT*rz**2*omT ))
    MassiVe = M1 + M2 - M3
        
#    Ip = np.array([np.sin(deltaP)/(dP)*Jp + 1.0j*JpPr*np.cos(deltaP), \
#                  (np.cos(deltaP)/(dP)*Jp - 1.0j*JpPr*np.sin(deltaP))*aP,
#                  (np.cos(deltaP)/(dP)*Jp - 1.0j*JpPr*np.sin(deltaP))*bP ] )
#    Im = np.array([np.sin(deltaM)/(dM)*Jm + 1.0j*JmPr*np.cos(deltaM), \
#                  (np.cos(deltaM)/(dM)*Jm - 1.0j*JmPr*np.sin(deltaM))*aM,
#                  (np.cos(deltaM)/(dM)*Jm - 1.0j*JmPr*np.sin(deltaM))*bM ] )
#    r = np.array([rx,ry,rz])
#    M33 = 2*np.real(np.dot(Im,np.conj(Ip))*np.dot(np.conj(Im),r)*np.dot(Ip,r) )
#    M11 = np.dot(Im,r)*np.conj(np.dot(Im,r))*np.dot(Ip,np.conj(Ip))
#    M22 = np.dot(Ip,r)*np.conj(np.dot(Ip,r))*np.dot(Im,np.conj(Im))
    return 2*np.pi**2*n**2*Rn*(Imag**2 + Rea**2 + 4*(n0/n)**(2)*MassiVe)
```

```python
def GammM(psi,Mpp,Mmm,n,n0):
    return (dblquad(lambda phi, theta: sin(theta)*\
         Gn(theta,phi,psi,Mpp,Mmm,n,n0), 0, np.pi, lambda x: 0, lambda x: 2.0*np.pi, epsabs=0.0, epsrel=0.01 )[0])
```
